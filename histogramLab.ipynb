## Lab 2: Split an image with 3 channels into one channel

```python
import numpy as np

# read the image
image2 = cv2.imread("imageSrc/lab2Image.jpg")

# convert the image to RGB
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(image2)
plt.title("Original Image")
plt.show()

# split the image into its three channels
(R, G, B) = cv2.split(image2)

# Display the R, G, B channels using matplotlib
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Red Channel - > gray")
plt.imshow(R, cmap='gray')

plt.subplot(1, 3, 2)
plt.title("Green Channel -> gray")
plt.imshow(G, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("Blue Channel -> gray")
plt.imshow(B, cmap='gray')

plt.show()
```
