# Praktikum 3 Geometric Transformations

## Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Scaling

```python
# Membaca gambar dari path img_path
img = cv2.imread("imageSrc/himmel2.jpeg")

# Melakukan scaling gambar dengan faktor 2x lebar dan 3x tinggi menggunakan interpolasi bicubic
height, width = img.shape[:2]
res = cv2.resize(img, (2*width, 3*height), interpolation=cv2.INTER_CUBIC)

# Mengonversi gambar input ke format RGB untuk ditampilkan dengan Matplotlib
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Mengonversi gambar hasil scaling ke format RGB untuk ditampilkan dengan Matplotlib
res_rgb = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)

# Menampilkan gambar asli dan hasil scaling secara terpisah
fig, axes = plt.subplots(1, 2, figsize=(12, 12)) # beda sama plt.subplot

# Menampilkan gambar asli di subplot kiri
axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")
axes[0].axis('off')

# Menampilkan gambar hasil scaling di subplot kanan
axes[1].imshow(res_rgb)
axes[1].set_title("Scaled Image (2x width & 3x height)")
axes[1].axis('off')

plt.show()
```

## Zoom in

```python
# Membaca gambar
image = cv2.imread("imageSrc/himmel.png")

# Menentukan faktor scaling untuk memperkecil gambar
factor = 2

# Menghitung dimensi baru berdasarkan faktor scaling
height, width = image.shape[:2]
new_height, new_width = int(height * factor), int(width * factor)

# Melakukan resize gambar dengan interpolasi linear
resized = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

# Mengonversi gambar input dan hasil zoom in dari BGR ke RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)

# Menampilkan gambar asli di subplot kiri
plt.figure(figsize=(12, 12))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")

# Menampilkan gambar hasil zoom in di subplot kanan
plt.subplot(1, 2, 2)
plt.imshow(resized_rgb)
plt.title("Zoomed-in Image")

plt.show()
```

## Zoom Out

```python
# Baca citra
img = cv2.imread("imageSrc/ayanokouji2.jpeg")

# Tentukan faktor perkecilan
factor = 2

# Hitung ukuran citra yang baru
height, width = img.shape[:2]
new_height, new_width = int(height / factor), int(width / factor)

# Perkecil citra menggunakan resize()
resized_img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_CUBIC)

# Mengonversi gambar input dan hasil zoom out dari BGR ke RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
resized_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB)

# Menampilkan gambar asli dan yang sudah di-zoom out
plt.figure(figsize=(12, 12))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(resized_img)
plt.title("Zoomed-out Image")

plt.show()
```

## Reflection

```python
# Membaca gambar
img = cv2.imread("imageSrc/ayanokouji.jpeg")


# Pencerminan citra menggunakan flip()
mirrored_img = cv2.flip(img, 1)

# Mengonversi gambar input dan hasil zoom out dari BGR ke RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
mirrored_img = cv2.cvtColor(mirrored_img, cv2.COLOR_BGR2RGB)

# Tampilkan citra asli dan citra yang sudah dipencerminan
plt.figure(figsize=(12, 12))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(mirrored_img)
plt.title("Mirrored Image")

plt.show()
```

## Translation

```python
# Membaca gambar dalam mode grayscale
image = cv2.imread("imageSrc/himmel2.jpeg", cv2.IMREAD_GRAYSCALE)

# Mengambil ukuran gambar
rows, cols = image.shape[:2]

# Membuat matriks transformasi untuk translasi
M = np.float32([[1, 0, 100], [0, 1, 50]])  # Translasi sebesar 100 piksel ke kanan dan 50 piksel ke bawah

# Melakukan translasi gambar menggunakan warpAffine
dst = cv2.warpAffine(image, M, (cols, rows))

# Mengonversi gambar input dan hasil translasi ke format RGB untuk ditampilkan dengan Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
dst_rgb = cv2.cvtColor(dst, cv2.COLOR_GRAY2RGB)

# Menampilkan gambar asli dan hasil translasi
plt.subplot(121), plt.imshow(image_rgb), plt.title("Gambar Asli")
plt.subplot(122), plt.imshow(dst_rgb), plt.title("Gambar Hasil Translasi")
plt.show()
```

## Rotation

```python
# Membaca gambar dalam mode grayscale
image = cv2.imread("imageSrc/histogram1.png", cv2.IMREAD_GRAYSCALE)

# Mengambil ukuran gambar
rows, cols = image.shape[:2]

# Membuat matriks transformasi untuk rotasi 90 derajat dengan pusat gambar
center = (cols / 2, rows / 2)   # mengambil koordinat tengah dari gambar
                                # untuk membuat matrix rotasi 2D
M = cv2.getRotationMatrix2D(center=center, angle=90, scale=1)

# Melakukan rotasi gambar menggunakan warpAffine
rotated_image = cv2.warpAffine(image, M, (cols, rows))

# Mengonversi gambar input dan hasil rotasi ke format RGB untuk ditampilkan dengan Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_GRAY2RGB)

# Menampilkan gambar asli dan hasil rotasi
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Gambar Asli")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Gambar Setelah Rotasi 90 Derajat")
plt.imshow(rotated_image_rgb)
plt.axis('off')

plt.show()
```

## Affine Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar berwarna
image = cv2.imread("imageSrc/lawliet1.jpeg")

# Mengambil ukuran gambar
rows, cols, ch = image.shape[:3]

# Menentukan titik-titik koordinat untuk transformasi afine
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])

# Menghitung matriks transformasi afine
M = cv2.getAffineTransform(pts1, pts2)

# Melakukan transformasi afine pada gambar
dst = cv2.warpAffine(image, M, (cols, rows))

# Mengonversi gambar input dan hasil transformasi ke format RGB untuk ditampilkan dengan Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
dst_rgb = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

# Menampilkan gambar asli dan hasil transformasi
plt.figure(figsize=(8, 5))

plt.subplot(1, 2, 1)
plt.title("Gambar Asli")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Gambar Setelah Transformasi Afine")
plt.imshow(dst_rgb)
plt.axis('off')

plt.show()
```

## Perspective Transformation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar berwarna
image = cv2.imread("imageSrc/jinwoo1.jpg")

# Mengambil ukuran gambar
rows, cols, ch = image.shape[:3]

# Menentukan titik-titik koordinat untuk transformasi perspektif
pts1 = np.float32([[50, 50], [200, 50], [50, 200], [200, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250], [200, 200]])

# Menghitung matriks transformasi perspektif
M = cv2.getPerspectiveTransform(pts1, pts2)

# Melakukan transformasi perspektif pada gambar
dst = cv2.warpPerspective(image, M, (cols, rows))

# Mengonversi gambar input dan hasil transformasi ke format RGB untuk ditampilkan dengan Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
dst_rgb = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)

# Menampilkan gambar asli dan hasil transformasi
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Gambar Asli")
plt.imshow(image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Gambar Setelah Transformasi Perspektif")
plt.imshow(dst_rgb)
plt.axis('off')

plt.show()
```

## Image Registration

```python
# Membaca dua gambar: satu gambar yang akan diperbaiki (align) dan satu sebagai referensi

image1 = cv2.imread('imageSrc/align.png')
image2 = cv2.imread('imageSrc/reference.png')
```

```python
# Mengonversi gambar ke grayscale
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Menggunakan ORB untuk mendeteksi dan mendeskripsikan fitur
orb = cv2.ORB_create()
keypoints1, descriptors1 = orb.detectAndCompute(gray1, None)
keypoints2, descriptors2 = orb.detectAndCompute(gray2, None)

# Menggunakan BFMatcher untuk mencocokkan fitur
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(descriptors1, descriptors2)

# Mengurutkan kecocokan berdasarkan jarak
matches = sorted(matches, key=lambda x: x.distance)

# Mengambil titik-titik yang cocok
points1 = np.zeros((len(matches), 2), dtype=np.float32)
points2 = np.zeros((len(matches), 2), dtype=np.float32)

for i, match in enumerate(matches):
    points1[i, :] = keypoints1[match.queryIdx].pt
    points2[i, :] = keypoints2[match.trainIdx].pt

# Menghitung homografi
H, mask = cv2.findHomography(points1, points2, cv2.RANSAC)

# Melakukan warp pada gambar yang akan diperbaiki
height, width, channels = image2.shape
aligned_image = cv2.warpPerspective(image1, H, (width, height))

# Mengonversi gambar input dan hasil transformasi ke format RGB untuk ditampilkan dengan Matplotlib
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
aligned_image_rgb = cv2.cvtColor(aligned_image, cv2.COLOR_BGR2RGB)

# Menampilkan gambar asli, referensi, dan hasil transformasi
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Gambar yang Akan Diperbaiki")
plt.imshow(image1_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Gambar Referensi")
plt.imshow(image2_rgb)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Gambar Setelah Align")
plt.imshow(aligned_image_rgb)
plt.axis('off')

plt.show()
```
